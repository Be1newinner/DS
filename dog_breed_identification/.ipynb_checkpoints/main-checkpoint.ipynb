## Preprocessing Images

```python
# 1. Converting images to numpy array

from matplotlib.pyplot import imread
image = imread(filenames[43])
# image.shape

# 2. turning image into tensors for GPU processing

tf.constant(image)
```

```python
# function to preprocess image

IMG_SIZE = 224

def process_image(image_path: str):
    """
    This will take an image path as input and convert into the tensor
    """

    image = tf.io.read_file(image_path) # 1. Reading the image file
    image = tf.image.decode_jpeg(image, channels=3) # 2. turning the image into numerical tensors with 3 channels
    image = tf.image.convert_image_dtype(image, tf.float32) # 3. converting the values in images from 0-255 into 0-1 which make gpu process faster
    image = tf.image.resize(image, size=[IMG_SIZE, IMG_SIZE]) # 4. Resizing image
    return image
```

```python
image = process_image(filenames[0])
# image
```

# 3. Converting tensors into batches

```python
# a. creating simple function to return a tuple (iamge, label)

def get_image_label(image_path, label):
    image = process_image(image_path)
    return (image, label)
```

```python
# get_image_label(filenames[42], boolean_labels[42])
```

```python
# b. Converting data into batches
BATCH_SIZE = 20

def create_data_batches(x, y=None, batch_size=BATCH_SIZE, valid_data=False, test_data=False):
    if test_data:
        print("creating test data batches")
        data = tf.data.Dataset.from_tensor_slices((tf.constant(x)))
        data_batch = data.map(process_image).batch(BATCH_SIZE)
        return data_batch
    elif valid_data:
        print("Creating validation data batches...")
        data = tf.data.Dataset.from_tensor_slices((tf.constant(x),tf.constant(y)))
        data_batch = data.map(get_image_label).batch(BATCH_SIZE)
        return data_batch
    else:
        print("Creating train data batches...")
        data = tf.data.Dataset.from_tensor_slices((tf.constant(x),tf.constant(y)))
        data = data.shuffle(buffer_size=len(x))
        data_batch =  data.map(get_image_label).batch(BATCH_SIZE)
        return data_batch
        
```

```python
train_data = create_data_batches(x_train, y_train)
val_data = create_data_batches(x_val, y_val, valid_data=True)
```

## Building the Image Recognition model

#### Before building the model we need
1. Input shape
2. Output Shape
3. URL of the Model

```python
INPUT_SHAPE = [IMG_SIZE, IMG_SIZE, 3] # batch , height , width , color channel

OUTPUT_SHAPE = len(unique_breeds) 
```

```python
# detector = hub.module_v2(Module)
# detector_output = detector(image_tensor, as_dict=True)
# class_names = detector_output["detection_class_names"]
```

```python
def create_model(input_shape=INPUT_SHAPE, output_shape = OUTPUT_SHAPE):
    print("Building model with MobileNetV2")
        
        
    base_model = tf.keras.applications.MobileNetV2(input_shape=input_shape,
                                                   include_top=False,
                                                   weights='imagenet')
    
    base_model.trainable = False
    
    # Add a global average pooling layer to flatten the output
    x = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)
    
    output_layer = tf.keras.layers.Dense(units=output_shape, activation='softmax')(x)
    
    model = tf.keras.Model(inputs=base_model.input, outputs=output_layer)
    
    model.compile(optimizer='adam',
                  loss=tf.keras.losses.CategoricalCrossentropy(),
                  metrics=['accuracy'])
    model.build(INPUT_SHAPE)
    
    return model
```

```python
model = create_model()

# model.summary()
```

```python
# Creating Callbacks
%load_ext tensorboard
```

```python
# TensorBoard logs

from datetime import datetime 

def create_tensorbaord_callback():
    # create a log directory

    logdir = os.path.join("logs", datetime.now().strftime("%Y%m%d-%H%M%S"))
    return tf.keras.callbacks.TensorBoard(logdir)

create_tensorbaord_callback()
```

```python
# Early Stopping

early_stopping = tf.keras.callbacks.EarlyStopping(
                                                monitor="val_accuracy",
                                                patience=3,
                                                restore_best_weights=True,
                                                verbose=1
                                                )
```

```python
# Training a model
NUM_EPOCHS = 100

def train_model():
    model = create_model()
    tensorboard = create_tensorbaord_callback()

    # Fit the model function
    model.fit(x=train_data,
             epochs=NUM_EPOCHS,
             validation_data=val_data,
             validation_freq=1,
             callbacks=[tensorboard, early_stopping] 
             )
    return model
```

```python
model = train_model()
```

```python
# Building function to get text label
def get_prediction_label(predection_probabilities):
    """
    Turns array of predections into labels 
    """
    return unique_breeds[np.argmax(predection_probabilities)]

# get_prediction_label(predictions[81])
```

```python
def unbatchify(data):
    """
    turns batches of (image, label) tensors into seperate arrays of images and labels
    """
    images = []
    labels = []

    for image, label in data.unbatch().as_numpy_iterator():
        images.append(image)
        labels.append(label)
    return images, labels

val_images, val_labels = unbatchify(val_data)
# val_images[0]
# val_labels[0]

get_prediction_label(val_labels[15])
```

```python
# Saving the tensor model

def save_model(model, suffix=None):
    """
    Saves a given model in the models directory and appneds a suffix ( string )
    """

    model_dir = os.path.join("models", datetime.now().strftime("%Y%m%d-%H%M%s"))
    model_path = model_dir + "_" + suffix + ".h5"
    print("model is saving to the path : ", model_path)
    model.save(model_path)
    return model_path
```

```python
# Loads a model

def load_model(model_path):
    """
    Loads the model from path
    """
    print("laoding the model")
    return tf.keras.models.load_model(model_path, custom_objects={"KerasLayer": hub.KerasLayer})
```
