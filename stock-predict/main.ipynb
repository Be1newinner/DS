```python
# Feature Engineering
```

```python
data = adani_data[["Date", "Open", "High", "Low", "Volume", "VWAP"]]

data.head()
```

```python
# Setting Index as Date
adani_data.set_index("Date", drop=False, inplace=True)
# adani_data.index
```

```python
adani_data.Close.plot(figsize=(14,7))
```

```python
adani_data.reset_index(drop=True, inplace=True)
lag_feature = ["Open", "High", "Low", "VWAP"]

window1 = 3
window2 = 7

ad_rolled_3d = adani_data[lag_feature].rolling(window=window1, min_periods=0)
ad_mean_3d = ad_rolled_3d.mean().shift(1).reset_index().astype(np.float32)

ad_rolled_7d = adani_data[lag_feature].rolling(window=window2, min_periods=0)
ad_std_7d = ad_rolled_7d.std().shift(1).reset_index().astype(np.float32)


for feaure in lag_feature:
    adani_data[f"{feaure}_mean_lag{window1}"] = ad_mean_3d[feaure].astype("float32")
    adani_data[f"{feaure}_std_lag{window2}"] = ad_std_7d[feaure].astype("float32")


# ad_mean_3d.head()
# adani_data.head()
```

```python
# Filling na

numeric_cols = adani_data.select_dtypes(include=["number"]).columns
adani_data[numeric_cols] = adani_data[numeric_cols].fillna(
    adani_data[numeric_cols].mean()
)

# print(adani_data[numeric_cols].mean())

# adani_data.isna().sum()
```

```python
# Adding Date as well and making date as index
adani_data.set_index("Date", drop=False, inplace=True)

adani_data["Date"] = pd.to_datetime(adani_data["Date"])
adani_data["Month"] = adani_data["Date"].dt.month
adani_data["day"] = adani_data["Date"].dt.day
adani_data["day_of_week"] = adani_data["Date"].dt.dayofweek

adani_data.head()
```

```python
# Removing Unnecessory columns

adani_data.drop(columns=["Symbol", "Series"], inplace=True)
adani_data.head()
```

### Model Building

```python
# Splitting data into Train and Testing Data

SPLIT_YEAR = "2018"

ad_train = adani_data[adani_data.Date < SPLIT_YEAR]
ad_valid = adani_data[adani_data.Date >= SPLIT_YEAR]

# ad_train.head()
# ad_valid.head()

print("Train shape ", ad_train.shape)
print("Valid shape ", ad_valid.shape)
print(f"Train Ratio :  {(ad_train.shape[0] / adani_data.shape[0] * 100):.2f}%")
print(f"Valid Ratio :  {(ad_valid.shape[0] / adani_data.shape[0] * 100):.2f}%")
```

```python
# splitting data into x and y

y_train = ad_train["Close"]
x_train = ad_train.drop(columns=["Close"], axis=1)

y_valid = ad_valid["Close"]
x_valid = ad_valid.drop(columns=["Close"], axis=1)

print(y_train.shape, x_train.shape)
print(y_valid.shape, x_valid.shape)
```
